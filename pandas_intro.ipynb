```python
def grade(mark):
    if mark < 6:
        return 'Троечник'
    elif mark >=6 and mark < 8:
        return 'Ударник'
    else:
        return 'Отличник'
```

```python
students['grade'] = students.loc[:, 'catps':'wpol'].mean(axis=1).apply(grade)
students['all_mean'] = students.loc[:, 'catps':'wpol'].mean(axis=1)
```

```python
students[['grade', 'all_mean']].head()
```

```python
students.groupby(['male'])[['wpol', 'compp']].agg(['mean', 'median', 'std'])
```
